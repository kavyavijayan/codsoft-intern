```python
num1 = int(input("Enter the first number:"))
num2 = int(input("Enter the second number:"))

operator = input("select any one(+,-,*,/):")

if(operator == '+'):
  print(num1 + num2)

elif (operator == '-'):
  print(num1 - num2)
elif(operator == '*'):
  print(num1 * num2)
else:
  print(num1 / num2)
```
